# TD Filtrage


## Système linéaire invariant

* Dans le cas de signaux à temps discrets, montrez que la sortie d'un système linéaire invariant
  s'exprime comme la convolution de l'entrée par la réponse impulsionnelle du système.
  Pour cela, on peut remarquer que l'entrée peut s’écrire comme une somme d’impulsions.
