## Produit de convolution

### Exercice 6

D'après la définition du produit de convolution,
et comme le signal $x$ est nul presque partout (sauf aux échantillons $n=1$ et $n=-1$),
alors :

$$
(x*y)[n]
&= \sum_{m=-\infty}^{+\infty} x[m] y[n-m] \\
&= x[-1] y[n+1] + x[1] y[n-1]
$$

Les valeurs de $x[-1]$ et $x[1]$ sont données par la définition de $x$, donc :

$$
(x*y)[n] = -y[n+1] + y[n-1].
$$ (E:rev:convolution)

D'après la définition de $y$, on a d'une part :

$$
y[n+1] =
\begin{cases}
  a &\text{pour } |n+1| \leq 1, \\
  0 &\text{ailleurs}
\end{cases} =
\begin{cases}
  a &\text{pour } -2 \leq n \leq 0, \\
  0 &\text{ailleurs}
\end{cases}
$$

et d'autre part :

$$
y[n-1] =
\begin{cases}
  a &\text{pour } |n-1| \leq 1, \\
  0 &\text{ailleurs}
\end{cases} =
\begin{cases}
  a &\text{pour } 0 \leq n \leq 2, \\
  0 &\text{ailleurs}
\end{cases}.
$$

D'après l'équation {eq}`E:rev:convolution`,
on peut finalement définir la valeur de tous les échantillons de $(x*y)$.
Il n'y a qu'en $n=0$ que les deux signaux $y[n+1]$ et $y[n-1]$ sont non nuls
(ils valent $a$).
Ainsi :

$$
(x*y) =
\begin{cases}
  -a &\text{pour } n = -2 \\
  -a &\text{pour } n = -1 \\
  0 &\text{pour } n = 0 \\
  a &\text{pour } n = 1 \\
  a &\text{pour } n = 2 \\
  0 &\text{ailleurs}
\end{cases}
$$
