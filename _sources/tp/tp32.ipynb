# ✐ Lissage par filtre moyenneur

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Chargement des données
data = np.loadtxt("../_static/files/temperatures.csv")
year = data[:,0]
temp = data[:,1]
N = len(year)
```

Réponse impulsionnelle du filtre moyenneur de taille 5 :

```python
h = np.ones(5)/5
smooth = np.convolve(temp, h, "same")
```

```python
# Affichage
plt.figure(figsize=(10,6))
plt.plot(year, temp, label="Mesures")
plt.plot(year, smooth, label="Lissage")
plt.xlabel("Année")
plt.ylabel("Température")
plt.legend()
plt.grid()
plt.show()
```

Pourquoi observe-t-on une décroissance des températures sur la courbe lissée aux alentours de 2020 ? Est-ce un signe que le dérèglement climatique est terminé ?
