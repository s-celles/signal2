# Filtres analogiques

Malgré le développement incontestable du traitement numérique du signal, le filtrage analogique reste indispensable.
Par exemple, les filtres anti-repliement seront incontournables pour réduire la bande passante d'un signal analogique.
De plus, certaines méthodes de synthèse de filtres numériques nécessitent de définir un filtre analogique.

Les premiers développements de filtres analogiques datent du début du XX<sup>e</sup> siècle avec la téléphonie.
Les premières théories ont été développées dans les années 1930–1940.
Les premiers filtres actifs (utilisant des composants actifs) sont apparus dans les années 1940.

## Synthèse des filtres analogiques

La méthodologie moderne pour synthétiser un filtre est appelée « _network synthesis_ ».

Dans le cadre du filtrage analogique,
c'est la tranformée de Laplace qui est utilisée pour étudier la fonction de transfert des filtres.

Comme on l'a vu avec la {ref}`Transformée de Laplace inverse <S:laplace:inverse>`,
on ne peut pas construire de filtre idéal.
Les fonctions de transfert des filtres analogiques classiques s'écrivent sous la forme d'une fraction rationnelle
et ne peuvent qu'approximer le gain d'un filtre idéal :

$$
H(s) = \frac{ \sum_{n=0}^N b_n s^n }{ \sum_{n=0}^M a_m s^m }
     = \frac{ b_Ns^N + b_{N-1}s^{N-1} + \dots + b_1s + b_0 }{ a_Ms^M + a_{M-1}s^{M-1} + \dots + a_1s + a_0 }
$$

L'enjeu revient donc à déterminer, en fonction du gabarit choisi, les valeurs des coefficients $b_n$ et $a_m$ :
c'est l'objet des travaux de [Butterworth](https://en.wikipedia.org/wiki/Stephen_Butterworth),
[Tchebychev](https://fr.wikipedia.org/wiki/Pafnouti_Tchebychev)
ou [Cauer](https://en.wikipedia.org/wiki/Wilhelm_Cauer).
Les caractéristiques de ces filtres varient donc en fonction de ces coefficients.

La plupart des méthodes de conception des filtres nécessitent de concevoir avant tout un « filtre passe-bas normalisé »,
c'est-à-dire que sa fréquence de coupure est égale à 1.
Nous verrons dans la suite l'expression des fonctions de transfert pour obtenir un filtre passe-bas normalisé.
À partir de la fonction de transfert $H(f)$ d'un filtre passe-bas normalisé,
on obtient un filtre quelconque à partir des transformations de fréquence ci-dessous :

* un filtre passe-bas de fréquence de coupure $f_c$ aura pour fonction de transfert :

  $$
  H\left(\frac{f}{f_c}\right)
  $$

* un filtre passe-haut de fréquence de coupure $f_c$ aura pour fonction de transfert :

  $$
  H\left(\frac{f_c}{f}\right)
  $$

* un filtre passe-bande de fréquences de coupure $f_{c-}$ et $f_{c+}$ aura pour fonction de transfert :

  $$
  H\left(\frac{f^2-f_{c-}f_{c+}}{f(f_{c+}-f_{c-})}\right)
  $$

* un coupe-bande de fréquences de coupure $f_{c-}$ et $f_{c+}$ aura pour fonction de transfert :

  $$
  H\left(\frac{f(f_{c+}-f_{c-})}{f^2-f_{c-}f_{c+}}\right)
  $$
  
Par la suite, nous ne traiteront donc que des filtres passe-bas.
Dans [ce document](https://vincmazet.github.io/signal2/_static/files/fana.pdf),
vous trouverez les réponses fréquentielles, réponses impulsionnelles et retard de groupe
pour les principaux filtres passe-bas normalisés analogiques.


### Filtre de Butterworth

Le gain d'un filtre de Butterworth passe-bas d'ordre $n$ est défini par :

$$
G(\omega) = \frac{1}{\sqrt{1+(\omega/\omega_c)^{2n}}}
$$

où $\omega = 2\pi f$ est la « pulsation » (d'unité rad.s<sup>–1</sup>)
et $\omega_c$ est la pulsation de coupure du filtre.

Le filtre de Butterworth est caractérisé par :
- aucune ondulation en bande passante,
- aucune ondulation en bande atténuée,
- une bande de transition assez large.


### Filtre de Bessel

Un filtre de Bessel passe-bas a pour fonction de transfert

$$
H(s) = \frac{\theta_n(0)}{\theta_n(s/\omega_c)}
$$

où $\omega_c$ est la pulsation de coupure du filtre
et $\theta_n(s)$ est le polynôme de Bessel tel que

$$
\theta_n(s) = \sum_{k=0}^n \frac{(2n-k)!}{2^{n-k} k! (n-k)!} s^k
\qquad\text{et}\qquad
\theta_n(0) = \lim_{x \to 0} \theta_n(x).
$$

Le filtre de Bessel est l'approximation d'un retard pur, et ses caractéristiques sont :
- un retard de groupe quasi-constant dans la bande passante,
- aucune ondulation en bande passante,
- aucune ondulation en bande atténuée,
- une bande de transition très large.



### Filtre de Tchebychev

Il existe deux types de filtre de Tchebychev (Chebyshev en anglais, Чебышёв en russe) : le type I et le type II.

Le gain d'un filtre passe-bas de Tchebychev de type I est défini par :

$$
G(\omega) = \frac{1}{\sqrt{ 1 + \varepsilon^2 T_n^2 \left(\frac{\omega}{\omega_c}\right) }}
$$

où $T_n$ est le [polynôme de Tchebychev](https://fr.wikipedia.org/wiki/Polyn%C3%B4me_de_Tchebychev) d'ordre $n$.

Le type I possède les caractéristiques suivantes :
- présence d'ondulations en bande passante,
- aucune ondulation en bande atténuée,
- bande de transition plus étroite que les filtres de Butterworth et Bessel,
- un retard de groupe très distordue près de la fréquence de coupure.

Le gain d'un filtre passe-bas de Tchebychev de type II est défini quant à lui par :

$$
G(\omega) = \frac{1}{\sqrt{ 1 + \frac{1}{ \varepsilon^2 T_n^2 \left(\frac{\omega_c}{\omega}\right) } }}
$$

Le type II possède les caractéristiques suivantes :
- aucune ondulation en bande passante,
- présence d'ondulations en bande atténuée,
- une bande de transition plus étroite que les filtres de Butterworth et Bessel.

### Filtre de Cauer

Le gain d'un filtre passe-bas de Cauer est défini par :

$$
G(\omega) = \frac{1}{\sqrt{1 + \varepsilon^2 R_n^2(\xi,\omega/\omega_c)}}
$$

où $R_n$ est la fonction elliptique d'ordre $n$, $\varepsilon$ est le taux d'ondulation et $\xi$ la sélectivité.

Celui-ci possède les caractéristiques ci-dessous :
* présence d'ondulations en bande passante,
* présence d'ondulations en bande atténuée,
* une transition très rapide,
* un retard de groupe très distordu près de la fréquence de coupure.
