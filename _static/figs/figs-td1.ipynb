```python
import matplotlib.pyplot as plt
import numpy as np
```

# Sinus cardinaux

```python
x = np.arange(-3, 3, 0.05)

data = {
    "a": np.sinc(x),
    "b": .22 + np.sinc(x),
    "c": np.absolute(np.sinc(x)),
    "d": 1.3 * (-.12 + np.sinc(x))
}

fig = plt.figure(figsize=(10,6))

for i, name in enumerate(data):

    ax = plt.subplot(2, 2, i+1)
    plt.plot(x, data[name], 'k', linewidth=1)
    plt.axis([-3, 3, -.5, 1.3])
    plt.text(-2.8, 1.2, name.upper(), fontsize=14, horizontalalignment='left', verticalalignment='top')
    plt.grid(True)

fig.savefig("td1-sinc.png", bbox_inches='tight')
```

# Signaux de convolution

```python
x = np.arange(-4, 4, 0.1)

data = {
    "dirac": [ [-4, 0, 0, 0, 4], [0, 0, 2, 0, 0] ],
    "diracs": [ [-4, -2, -2, -2, 2, 2, 2, 4], [0, 0, 2, 0, 0, 2, 0, 0] ],
    "narrow-door": [ [-4, -.5, -.5, .5, .5, 4], [0, 0, 1, 1, 0, 0] ],
    "large-door": [ [-4, -2, -2, 2, 2, 4],  [0, 0, 1, 1, 0, 0] ],
    "triangle": [ [-4, -2, 0, 2, 4], [0, 0, 2, 0, 0] ],
    "gaussian": [ x, 2*np.exp(-x*x/0.5) ]
}

convs = [
    ["dirac", "triangle"],
    ["diracs", "gaussian"],
    ["narrow-door", "large-door"],
]

for j, c in enumerate(convs):
    
    fig = plt.figure(figsize=(10,2))
    for i, name in enumerate(c):

        ax = plt.subplot(1, 2, i+1)
        plt.plot(data[name][0], data[name][1], 'k', linewidth=1)
        if i==0:
            plt.text(4.5, 1, "*", fontsize=20)
        plt.grid(True)

        plt.axis([-4, 4, -.5, 3])

    fig.savefig(f"td1-conv-{j+1}.png", bbox_inches='tight')
```
